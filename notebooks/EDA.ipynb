Importing vehicles_us.csv and performing basic EDA.  

```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('../vehicles_us.csv')

print(df.head())
print(df.info())
print(df.describe())
df.sample(10)
```

```python
df['odometer'].max()
```

```python
df['model_year'] = df['model_year'].astype('object')
df['cylinders'] = df['cylinders'].astype('object')
df['is_4wd'] = df['is_4wd'].astype("object")
df.dtypes
```

```python
px.histogram(df[df['price'] <= 100000], x='price', nbins=75, title='Distribution of Vehicle Prices')
```

From the histogram, it appears that most used cars are listed for less than $50,000, with most being listed between $4,000 and $6,000.  The histogram filters out cars priced over $100K, since those listings are outliers.

```python
px.scatter(df, x="odometer", y="price", color="condition", title="Price vs. Mileage by Condition")
```

The scatter plot above shows that as mileage increases, both price and number of cars listed decrease

```python
# Group and compute mean and count
df_grouped = df.groupby("type").agg(avg_price=("price", "mean"), count=("price", "count")).reset_index()

# Sort by count
df_grouped = df_grouped.sort_values(by="count", ascending=False)

# Plot average price sorted by count
px.bar(df_grouped, x="type", y="avg_price", title="Average Price by Vehicle Type (Sorted by Count)")
```

The above bar plot shows the average price of vehicle type sorted by number of vehicles listed.  The most popular vehicle type listed is SUV at an average price of $11,149.  Trucks are the second most popular vehicle listed at an average price of $16,735.

```python
px.histogram(df, x="fuel", color="type", barmode="group", title="Fuel Types by Vehicle Type")
```

The vast majority of vehicles listed require gasoline as a fuel soource, followed by diesel, hybrid, other, and electric.  

```python
px.scatter(df, x="days_listed", y="price", color="condition", title="How Price Relates to Time on Market")
```

As vehicles spend more time on the market, the listed price for each vehicle tends to decrease.  

```python
df['is_4wd'] = df['is_4wd'].fillna(0).astype(object)
px.box(df, x="is_4wd", y="days_listed", title="Distribution of Time on Market by 4WD Status", labels={"is_4wd": "4WD (0 = No, 1 = Yes)", "days_listed": "Days Listed"})
```

Vehicles listed with 4WD spend about the same amount of time on the market as vehicles lacking 4WD.
